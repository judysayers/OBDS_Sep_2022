```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
counts = pd.read_csv('2.2a_ex1_featurecounts.tsv', sep='\t', comment= '#', header=0)
```

```python
counts.columns
```

```python
counts = counts.drop(['Chr', 'Start', 'End', 'Strand', 'Length'], axis=1) 
```

```python
counts
counts.columns
```

```python
counts = counts.rename(columns={'bam/RS411-nasc-1.bam':'RS411-nasc-1','bam/RS411-nasc-2.bam':'RS411-nasc-2','bam/RS411-nasc-3.bam':'RS411-nasc-3','bam/SEM-nasc-1.bam':'SEM-nasc-1','bam/SEM-nasc-2.bam':'SEM-nasc-2','bam/SEM-nasc-3.bam':'SEM-nasc-3'})
counts
```

```python
counts['row_sum'] = counts.sum(axis = 1)
counts
```

```python
counts = counts.query("row_sum > 100")
counts
```

```python
counts.columns
```

```python
sns.pairplot(counts.loc[:, ['RS411-nasc-1', 'RS411-nasc-2', 'RS411-nasc-3',
       'SEM-nasc-1.bam', 'SEM-nasc-2', 'SEM-nasc-3']], kind = 'scatter')
```

```python
counts
```

```python
counts_2 = counts.loc[:, ['Geneid', 'RS411-nasc-1', 'RS411-nasc-2', 'RS411-nasc-3',
       'SEM-nasc-1.bam', 'SEM-nasc-2', 'SEM-nasc-3']]
```

```python
counts_2 = counts_2.melt(id_vars=["Geneid"], var_name = "cell_replicate", value_name = "count_total")
counts_2
```

```python
sns.histplot(data=counts_2, x="count_total", hue="cell_replicate", bins = 1000)
plt.xlim(0, 5000)
```

```python
sns.kdeplot(data=counts_2, x="count_total", hue="cell_replicate", palette="husl") 
plt.xlim(0, 5000)
```

```python
sns.violinplot(data=counts_2, y="count_total", x="cell_replicate", palette = "flare", as_cmap=True)
plt.ylim(0,7500)
```

```python
counts_2 = counts_2.set_index("Geneid")
highest_genes = counts_2.groupby("cell_replicate")["count_total"].nlargest(10)
```

```python
highest_genes.reset_index()
```

```python
counts_2["replicate_sum"] = counts_2.groupby("cell_replicate").transform(np.sum)
```

```python
counts_2
```

```python
counts_2["normalised_counts"] = counts_2["count_total"] / counts_2["replicate_sum"]
```

```python
counts_2
```

```python
counts_2["normalised_CPM"] = counts_2["normalised_counts"] * 1e6
```

```python
counts_2
```

```python
counts_2["log_norm_CPM"] = np.log10(counts_2["normalised_CPM"])
```

```python
counts_2
```

```python
#counts_2["log_CPM_var"] = 
counts_2.groupby("Geneid").agg(np.var(counts_2["log_norm_CPM"]))
```
